```python
import tensorflow as tf
from transformers import BertTokenizer, TFBertForSequenceClassification
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=3)
```

```python
texts = ["Hello world!", "How are you?", "BERT is amazing."]
labels = [0, 1, 2]
```

```python
inputs = tokenizer(texts, padding=True, truncation=True, max_length=140, return_tensors='tf')
labels = tf.convert_to_tensor(labels)
```

```python
dataset = tf.data.Dataset.from_tensor_slices((dict(inputs), labels))
dataset = dataset.batch(2)
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=2e-5),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(dataset, epochs=3)
```

```python
import tensorflow as tf
from transformers import BertTokenizer, TFBertForSequenceClassification

# Cargar el modelo y tokenizador
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=3)

# (El resto de tu código para el entrenamiento aquí...)

# Exportar el modelo para TensorFlow Serving
@tf.function(input_signature=[tf.TensorSpec([None, 140], tf.int32, name="input_ids"),
                               tf.TensorSpec([None, 140], tf.int32, name="attention_mask")])
def serving_fn(input_ids, attention_mask):
    return model(input_ids, attention_mask=attention_mask)

# Guardar el modelo exportado
model.save('mybert/1', save_format='tf', signatures={'serving_default': serving_fn})
```
